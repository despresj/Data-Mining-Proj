```python
predictions = predictions.rename(columns={"Unnamed: 0":"date"})
preds = predictions.iloc[:, 3:]
predictions["week"] = pd.DatetimeIndex(predictions.date).week
predictions["month"] = pd.DatetimeIndex(predictions.date).month
predictions["year"] = pd.DatetimeIndex(predictions.date).year
```

```python
predictions
```

```python
sum_preds = predictions\
    .groupby(["week", "month", "year", "store_item"]).sum()
```

```python
def smape(F, A = sum_preds.sales):
    return 100/len(A) * np.sum(2 * np.abs(F - A) / (np.abs(A) + np.abs(F)))
```

```python
sum_preds.apply(smape).sort_values().round(2)
```
