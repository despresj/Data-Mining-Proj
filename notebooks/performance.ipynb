```python
import pandas as pd

errors = pd.read_csv("../data/predictions/errors.csv")
mape = errors.iloc[:, 3:].sub(errors.sales, axis=0)
errors["month"] = pd.DatetimeIndex(errors.date).month
errors["year"] = pd.DatetimeIndex(errors.date).year

mape\
    .abs()\
    .mean()\
    .sort_values()\
    .round(2)
```

```python
pd.concat([errors[["date", "store_item", "month", "year"]], mape], axis=1)
```

```python
pd.concat([errors[["date", "store_item", "month", "year"]], mape], axis=1)\
    .groupby(["month", "year", "store_item"])\
        .sum()
```
